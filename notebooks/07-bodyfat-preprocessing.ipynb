```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

plt.hist(df["BodyFat"], bins=20)

plt.title("Distribution of Body Fat")

plt.xlabel("Body Fat (%)")

plt.ylabel("Number of People")

plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.hist(df["Age"], bins=20)

plt.title("Age Distribution")

plt.xlabel("Age")

plt.ylabel("Count")

plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.hist(df["Weight"], bins=20)

plt.title("Weight Distribution")

plt.xlabel("Weight")

plt.ylabel("Count")

plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.hist(df["Height"], bins=20)

plt.title("Height Distribution")

plt.xlabel("Height")

plt.ylabel("Count")

plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.scatter(df["Weight"], df["BodyFat"])

plt.xlabel("Weight")

plt.ylabel("Body Fat")

plt.title("Weight vs Body Fat")

plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.scatter(df["Height"], df["BodyFat"])

plt.xlabel("Height")

plt.ylabel("Body Fat")

plt.title("Height vs Body Fat")

plt.grid(True)

plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(12,10))

sns.heatmap(
    df.corr(),
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Matrix")

plt.show()
```

```python
sns.pairplot(
    df[
        [
            "Age",
            "Weight",
            "Height",
            "BodyFat"
        ]
    ]
)

plt.show()
```
